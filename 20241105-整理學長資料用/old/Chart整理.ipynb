```python
# 變數設定，目的在於其他的project只要修改這部分就好了
Project = 'chart'
Project2 = 'Chart'
project_bugid_num = 26
SBFL_tool = 'Ample'

# 這是要放要拿來使用的絕對路徑
unsort_data_path = 'Z:/廷偉/偉/研究所/其他SBFL資料-用於後續實驗用/CalculateSpectrum/CalculateSpectrum/CalculateSpectrum/Chart/Sus/old'
```

```python
# 取得project的每一個bug-id有多少行的方法。
SBFL_fault_statement_number = []
project_statement_file = 'project_line/' + Project +'_line.txt'

# 使用 with 語句打開文件，確保文件在使用後自動關閉
with open(project_statement_file, 'r', encoding='utf-8') as file:
    project_lines = file.readlines()  # 讀取所有行

# 去掉每行末尾的換行符
project_lines = [int(line.strip()) for line in project_lines] # 順便將str轉為數字
print(project_lines)

max_line = max(project_lines) # 找出最大行數，用於後續輸出使用
print(max_line)
```

```python
# get_fault_statement
#動態產生二微陣列的方法
Project_bugid_faultStatement = []

for i in range(0,project_bugid_num):
    fault_statement_file_path = '錯誤位置/'+ Project + '/' + Project2 +'_v' + str(i+1) + '_FaultStatement.txt'
    print(fault_statement_file_path)

    # 使用 with 語句打開文件，確保文件在使用後自動關閉
    with open(fault_statement_file_path, 'r', encoding='utf-8') as file:
        lines2 = file.readlines()  # 讀取所有行

    # 去掉每行末尾的換行符
    lines2 = [line.strip() for line in lines2]
    print(lines2)

    lines2 = lines2[0].split()
    print(len(lines2))

    Project_bugid_faultStatement.append([])
    for j in range(0,len(lines2)):
        Project_bugid_faultStatement[i].append(int(lines2[j])) # 順便將str轉為數字

print(Project_bugid_faultStatement)

```

```python
import os

for i in range(0,project_bugid_num):
    #file_path = Project + '_test/' + SBFL_tool + '_v' + str(i+1) + '_sus.txt' #使用相對路徑
    file_path = unsort_data_path + '/' + SBFL_tool + '_v' + str(i+1) + '_sus.txt' #使用絕對路徑

    print(file_path)

    output_file2 = Project + '_result/' + SBFL_tool +'_' + Project2 + '-' + str(i+1) + '_result'# 輸出的懷疑直Top結果，檔案要存在哪一個資料夾，檔案名稱要存成甚麼。

    # 使用 with 語句打開文件，確保文件在使用後自動關閉
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # 讀取所有行

    # 去掉每行末尾的換行符
    lines = [line.strip() for line in lines]

    # 將列表中的元素和其索引配對，用於表示的幾個statement
    indexed_list = list(enumerate(lines))

    # 將列表中的元素和其索引配對,索引從1開始
    indexed_list = [(i+1, line) for i, line in enumerate(lines)]

    # 找出前 XX 名（若列表長度小於 XX，則返回所有元素）
    top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:max_line]

    # 將輸出結果寫入新的文件
    open(output_file2 + '.txt', 'w').close()
    Now_top = 0 # 用於紀錄top用的
    previous_rank_sus = '' #表示與前一個rank的懷疑質做比較做比較
    count_top = 0
    for index, value in top_test:

        #用於處理計算Top使用----
        count_top += 1
        if value != previous_rank_sus:
            #Now_top = Now_top
            #Now_top += 1
            Now_top = count_top
        else:
            Now_top = Now_top
        #----------------------

        previous_rank_sus = value
        f = open(output_file2 + '.txt', 'a')
        f.write(f"Top: {Now_top}, Spectra_file_index: {index}, Suspicious_Value: {value}\n")

        f.close() # 要記得close，不然沒辦法做後續處理。
```
