```python

for SBFL_index, SBFL_a in enumerate(SBFL_all):
    #要存成的excel index有甚麼
    data = {
        'SBFL_Tool': [],
        'bug_id': [],
        'Top': [],
        'Spectra_file_index': [],
        'Suspicious_Value': []
    }
    for i in range(0,project_bugid_num):
        file_path = unsort_data_path + '/' + SBFL_a + '_v' + str(i+1) + '_sus.txt' #使用絕對路徑
        print(file_path)

        output_file2 = Project + '_result/' + SBFL_a +'_' + Project2 + '-' + str(i+1) + '_result'# 輸出的懷疑直Top結果，檔案要存在哪一個資料夾，檔案名稱要存成甚麼。

        # 使用 with 語句打開文件，確保文件在使用後自動關閉
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()  # 讀取所有行

        # 去掉每行末尾的換行符
        lines = [line.strip() for line in lines] 

        # 轉換為float，以及處理例外的問題。
        numbers = []
        for line in lines:
            try:
                number = float(line.strip())
            except ValueError:  # 捕獲轉換錯誤
                print("抓到非數字!")
                number = 0  # 將無法轉換的值設為0
            numbers.append(number)

        # 將列表中的元素和其索引配對，用於表示的幾個statement
        indexed_list = list(enumerate(numbers))

        # 取出最大的數字
        max_prob = max(indexed_list)
        count = numbers.count(max_prob)

        # 將列表中的元素和其索引配對,索引從1開始
        indexed_list = [(i+1, numbers) for i, numbers in enumerate(numbers)]

        print(len(indexed_list))

        # 找出前 XX 名（若列表長度小於 XX，則返回所有元素）
        # top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:max_line]

        # 找出前 200 名（若列表長度小於 200，則返回所有元素），因為全部都存起來，excel會爆炸，所以改成200名就好了。
        top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:200]
        
        Now_top = 0 # 用於紀錄top用的
        previous_rank_sus = '' #表示與前一個rank的懷疑質做比較做比較
        count_top = 0

        for index, value in top_test:

            #用於處理計算Top使用----
            count_top += 1
            if value != previous_rank_sus:
                #Now_top = Now_top
                #Now_top += 1
                Now_top = count_top
            else:
                Now_top = Now_top
            #----------------------

            previous_rank_sus = value
            data['SBFL_Tool'].append(SBFL_a)
            data['bug_id'].append(i+1)
            data['Top'].append(Now_top)
            data['Spectra_file_index'].append(index)
            data['Suspicious_Value'].append(value)

    # 創建 DataFrame
    df = pd.DataFrame(data)
    print(df)
    print(len(df))

    save_excel_file_name = Excel_name + SBFL_a
    print(save_excel_file_name)
    wb = openpyxl.Workbook() #建立空的excel
    #wb = openpyxl.load_workbook('oxxo.xlsx') #開啟excel
    wb.save( Excel_save_path + '/' + save_excel_file_name + '.xlsx')

    # # 將資料寫入 Excel 檔案
    file_name = Excel_save_path + '/' + save_excel_file_name + '.xlsx'
    df.to_excel(Excel_save_path + '/' + save_excel_file_name + '.xlsx', index=False)

    print(f"資料已寫入 {save_excel_file_name} 文件中")
```
