```python
# print(df.index)
# print(df.columns)
# print(df.loc[:,1:3])

# 先建立一個空的一維陣列
Top_1 = []
Top_3 = []
Top_5 = []
Top_10 = []

# 查看 DataFrame 當前的記憶體用量
#df.info(memory_usage="deep")

# print(df.loc[0].values)
for SBFL_Tool_index in range(0,len(SBFL_all)):

    # 每個SBFL都要建立一個二維
    Top_1.append([])
    Top_3.append([])
    Top_5.append([])
    Top_10.append([])
    #-----------------------------

    # 每個SBFL都要建立一個dataframe
    excel_path = Project + '_result/' + Project + '_整理結果' + SBFL_all[SBFL_Tool_index] + '.xlsx'
    print(excel_path)
    df = pd.read_excel(excel_path, engine='openpyxl', sheet_name='Sheet1') #輸入要用的project excel是甚麼
    print(df)
    #-----------------------------

    excel_index = 0 #紀錄目前的SBFL工具，bugid會用掉多少行
    for i in range(0,project_bugid_num):
        print(project_lines[i])

        for bugid_index in range(0, int(project_lines[i])):
            #create_df
            find_flag = 0 # 表示有沒有找到
            get_row_data = df.loc[bugid_index + excel_index].values # 取出目前row的資料
            #print("get_row_data: ", get_row_data)
            #print(df.loc[bugid_index + excel_index].values) 
            for faultStatement_index in range(0,len(Project_bugid_faultStatement[i])): # 每個錯誤位置都要檢查。
                #print(get_row_data)
                if get_row_data[2] == 1 and get_row_data[3] == Project_bugid_faultStatement[i][faultStatement_index]:
                    Top_1[SBFL_Tool_index].append(i+1)
                    Top_3[SBFL_Tool_index].append(i+1)
                    Top_5[SBFL_Tool_index].append(i+1)
                    Top_10[SBFL_Tool_index].append(i+1)
                    print("Top_1")
                    find_flag = 1
                    break #跳出for
                elif 1 < get_row_data[2] <= 3 and get_row_data[3] == Project_bugid_faultStatement[i][faultStatement_index]:
                    Top_3[SBFL_Tool_index].append(i+1)
                    Top_5[SBFL_Tool_index].append(i+1)
                    Top_10[SBFL_Tool_index].append(i+1)
                    print("Top_3")
                    find_flag = 1
                    break #跳出for
                elif 3 < get_row_data[2] <= 5 and get_row_data[3] == Project_bugid_faultStatement[i][faultStatement_index]:
                    Top_5[SBFL_Tool_index].append(i+1)
                    Top_10[SBFL_Tool_index].append(i+1)
                    print("Top_5")
                    find_flag = 1
                    break #跳出for
                elif 5 < get_row_data[2] <= 10 and get_row_data[3] == Project_bugid_faultStatement[i][faultStatement_index]:
                    Top_10[SBFL_Tool_index].append(i+1)
                    print("Top_10")
                    find_flag = 1
                    break #跳出for
                elif get_row_data[2] > 10:
                    find_flag = 1
                    print("Not Found!")
                    break #跳出for
            if find_flag == 1:
                break #找到後跳出
        excel_index = excel_index + int(project_lines[i]) # bugid完成後，excel的index跳到下一個bugid的index。

        Now_message = "Now bugid_" + str(i+1) + " end excel index: "
        print( Now_message, excel_index)
        #print(df.loc[excel_index].values)
        print("Top_1: " , Top_1[SBFL_Tool_index])
        print("Top_3: " , Top_3[SBFL_Tool_index])
        print("Top_5: " , Top_5[SBFL_Tool_index])
        print("Top_10: " , Top_10[SBFL_Tool_index])
        print("-----------------------------------")
    df = pd.DataFrame() # 清空dataframe
```
