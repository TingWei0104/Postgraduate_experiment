```python
import os

for i in range(0,project_bugid_num):
    #print(i+1)
    file_path = Project + '_test/' + SBFL_tool + '_v' + str(i+1) + '_sus.txt'
    print(file_path)

    output_file2 = Project + '_result/' + Project2 +'-' + str(i+1) + '_' + SBFL_tool + '_result'# 輸出的懷疑直Top結果，檔案要存在哪一個資料夾，檔案名稱要存成甚麼。

    # 使用 with 語句打開文件，確保文件在使用後自動關閉
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # 讀取所有行

    # 去掉每行末尾的換行符
    lines = [line.strip() for line in lines]

    # 將列表中的元素和其索引配對，用於表示的幾個statement
    indexed_list = list(enumerate(lines))

    # 將列表中的元素和其索引配對,索引從1開始
    indexed_list = [(i+1, line) for i, line in enumerate(lines)]

    # 找出前 XX 名（若列表長度小於 XX，則返回所有元素）
    top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:max_line]

    #print(top_test)
    #print(top_test[0])

    # 將輸出結果寫入新的文件
    open(output_file2 + '.txt', 'w').close()
    Now_top = 0 # 用於紀錄top用的
    previous_rank_sus = '' #表示與前一個rank的懷疑質做比較做比較
    for index, value in top_test:

        #用於處理計算Top使用----
        if value != previous_rank_sus:
            Now_top = Now_top
            Now_top += 1
        else:
            Now_top = Now_top
        #----------------------

        previous_rank_sus = value
        f = open(output_file2 + '.txt', 'a')
        f.write(f"Top: {Now_top}, Stetemnet: {index}, Suspicious_Value: {value}\n")

        f.close() # 要記得close，不然沒辦法做後續處理。
```
