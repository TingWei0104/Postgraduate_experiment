```python
my_list = [10, 20, 15, 40, 50, 5, 30, 25, 45, 35, 60]  # 示例列表

# 將列表中的元素和其索引配對
indexed_list = list(enumerate(my_list))

# 找出前 50 名（若列表長度小於 50，則返回所有元素）
top_50 = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:3]

# 輸出結果
for index, value in top_50:
    print(f"Index: {index}, Value: {value}")
```

```python
import os

for i in range(0,26):
    #print(i+1)
    file_path = 'chart_test/Ample_v' + str(i+1) + '_sus.txt'
    print(file_path)

    output_file2 = 'chart_result/Chart-' + str(i+1) + '_Ample_result'# 輸出的懷疑直Top結果，檔案要存在哪一個資料夾，檔案名稱要存成甚麼。

    # 使用 with 語句打開文件，確保文件在使用後自動關閉
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # 讀取所有行

    # 去掉每行末尾的換行符
    lines = [line.strip() for line in lines]

    # 將列表中的元素和其索引配對，用於表示的幾個statement
    indexed_list = list(enumerate(lines))

    # 將列表中的元素和其索引配對,索引從1開始
    indexed_list = [(i+1, line) for i, line in enumerate(lines)]

    # 找出前 XX 名（若列表長度小於 XX，則返回所有元素）
    top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:100]

    # 將輸出結果寫入新的文件
    open(output_file2 + '.txt', 'w').close()
    for index, value in top_test:
        f = open(output_file2 + '.txt', 'a')
        f.write(f"Stetemnet: {index}, Suspicious_Value: {value}\n")

        f.close() # 要記得close，不然沒辦法做後續處理。

    print(f"Output saved to {output_file2}") # 顯示用，可忽略。

    # 單純輸出結果
    # print("現在是" + file_path + ": ")
    # for index, value in top_test:
    #     print(f"statement_num: {index}, Value: {value}")
```
