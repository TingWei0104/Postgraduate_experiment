```python
#用於最後寫入用
data = {
    'bugid': [],
    'combine': [],
    'top-1': [],
    'top-3': [],
    'top-5': [],
    'top-10': [],
    '是否生成(0)/正確(0.5, 1)': [],
    'correct patch/lable': [],
    '標示用，不參與訓練': [],
    '補充說明(如果有的話)': []
}
#print(data['bugid'][0])
```

```python
for i in range(1, bugid_num+1): #chart共有26個
    combine = 1
    for SBFL_index, SBFL in enumerate(SBFLlist):
        for APR_index, APR in enumerate(APRlist):
            data['bugid'].append(i+id_start)#2024/10/22，已將bug-id編號修改，time由40001開始到40027
            data['combine'].append(combine)
            combine = combine + 1
            #print(APR_index)
            #print(str(APR_index) + ': ' + SBFL + ', ' + APR)
            #處理Top-1
            if Top1[SBFL_index][i-1] == 1:
                data['top-1'].append(1)
            elif Top1[SBFL_index][i-1] == 0:
                data['top-1'].append(0)
            #處理Top-3
            if Top3[SBFL_index][i-1] == 1:
                data['top-3'].append(1)
            elif Top3[SBFL_index][i-1] == 0:
                data['top-3'].append(0)
            #處理Top-5
            if Top5[SBFL_index][i-1] == 1:
                data['top-5'].append(1)
            elif Top5[SBFL_index][i-1] == 0:
                data['top-5'].append(0)
            #處理Top-10
            if Top10[SBFL_index][i-1] == 1:
                data['top-10'].append(1)
            elif Top10[SBFL_index][i-1] == 0:
                data['top-10'].append(0)
            #標示是否正確，以及label
            if APR_result[APR_index][i-1] == 0:
                data['是否生成(0)/正確(0.5, 1)'].append(0)
                data['correct patch/lable'].append(0)
            elif APR_result[APR_index][i-1] == 1:
                data['是否生成(0)/正確(0.5, 1)'].append(1)
                data['correct patch/lable'].append(0)
            elif APR_result[APR_index][i-1] == 2:
                data['是否生成(0)/正確(0.5, 1)'].append(2)
                data['correct patch/lable'].append(1)
            data['標示用，不參與訓練'].append(project_is + '-' + str(i) + ', ' + SBFL + ', ' + APR + ', combine: ' + str(combine-1))

            
            data['補充說明(如果有的話)'].append(0)


print(data['bugid'])
print(len(data['bugid']))
print(data['combine'])
print(len(data['combine']))
print(data['top-1'])
print(len(data['top-1']))
print(data['top-3'])
print(len(data['top-3']))
print(data['top-5'])
print(len(data['top-5']))
print(data['top-10'])
print(len(data['top-10']))
print(data['是否生成(0)/正確(0.5, 1)'])
print(len(data['是否生成(0)/正確(0.5, 1)']))
print(data['correct patch/lable'])
print(len(data['correct patch/lable']))
print(data['標示用，不參與訓練'])
print(len(data['標示用，不參與訓練']))
print(data['補充說明(如果有的話)'])
print(len(data['補充說明(如果有的話)']))
            
```

```python
# 創建 DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
#將資料normalize使用

#複製一個df，將其進行normalize
df_max_scaled = df.copy()

# 需要標準化的特徵列
# 在前面資料準備有補上了，所以註解掉。
#columns_to_normalize = ['top-1', 'top-3', 'top-5', 'top-10', '是否生成(0)/正確(1, 2)']

# 應用標準化技術
for column in columns_to_normalize:
    df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()

# 查看標準化後的數據
display(df_max_scaled)
```
