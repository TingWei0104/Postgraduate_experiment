```python
# 建立紀錄APR使用的list
acs = []
Nopol = []
SimFix = []
TBar = []
jGenProg = []
jKali = []
jMutRepair = []

#用於最後寫入用
excel_data = {
    'project': [], # 0
    'bugid': [], # 1
    'APR': [], # 2
    'label': []  # 3
}
```

```python
# 計算APR在這個patch總共有多少的bug_id有產生plausible patch。
now_list_index2 = 0
APR_patch = []
for APR_index in range(APR_num): # 0:ACS、1:Nopol、2:SimFix、3:TBar、4:jGenProg、5:jKali、6:jMutRepair
    patch_num = 0
    for now_bugid in range(project_bugid):
        now_indexx = bug_id_list_index[now_bugid] + APR_index
        #print(data_list[now_indexx])
        if data_list[now_indexx][6] == 1:
            patch_num += 1
    APR_patch.append(patch_num)

print(APR_patch)
```

```python
list_index = len(data_list)
now_list_index = 0
for now_bugid in range(project_bugid):
    # 表示目前的bug_id從哪個index開始
    now_list_index = bug_id_list_index[now_bugid]
    for APR_index in range(APR_num): # 0:ACS、1:Nopol、2:SimFix、3:TBar、4:jGenProg、5:jKali、6:jMutRepair

        #print("Now is ", project_is, '_', now_bugid + 1, ' ,', APRlist[APR_index])
        caculate_index = 0 + now_list_index + APR_index # APR開始的index
        print("caculate_index: ", caculate_index)
        molecular = 0 # 分子
        for SBFL_index in range(SBFL_num):
            #print("Now is ", project_is, '_', now_bugid + 1, ' ,', APRlist[APR_index])
            molecular = molecular + data_list[caculate_index][5] # 計算所有這個APR的SBFL分子

            # 20250210 考慮中
            # if APR_result[APR_index][SBFL_index] == 0:
                

            caculate_index = caculate_index + 7 # 下一個相同的APR就是再加7個index，計算完成再加index不然會超出range
            
        
        
        caculate_index = 0
        label = molecular / SBFL_num 

        # 考慮有沒有需要
        label = label * (APR_patch[APR_index] / project_bugid) # 避免同一個bug_id，使用這個計算方法，7個APR所產生的數值都是一樣的
        

        #print("label: ", label)
        if APR_index == 0:
            acs.append(label)
        elif APR_index == 1:
            Nopol.append(label)
        elif APR_index == 2:
            SimFix.append(label)
        elif APR_index == 3:
            TBar.append(label)
        elif APR_index == 4:
            jGenProg.append(label)
        elif APR_index == 5:
            jKali.append(label)
        elif APR_index == 6:
            jMutRepair.append(label)
        
        excel_data['project'].append(project_is)
        excel_data['bugid'].append(now_bugid + 1)
        excel_data['APR'].append(APRlist[APR_index])
        excel_data['label'].append(label)
```
