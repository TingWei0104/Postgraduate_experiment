```python
#用於將S1~S5整理用的

import os
import shutil

pre_fold_file_name = ['S1.txt', 'S2.txt', 'S3.txt', 'S4.txt', 'S5.txt']
fold_foder = ['Fold1', 'Fold2', 'Fold3', 'Fold4', 'Fold5']
for i in range(5):
    file_names= list(pre_fold_file_name) #為了不更動到原本list，將list複製。
    source_file = file_names[i] #作為test set的檔案
    #print(source_file)
    file_names.pop(i) #把作為test set的檔案去除
    #print(file_names) #驗證集依序為1->2->3->4->5。
    output_folder = fold_foder[i] #要存的資料夾
    #print(output_folder)
    output_file = os.path.join(output_folder, 'training.txt') #training set的檔案名稱
    destination_file = os.path.join(output_folder, 'test.txt') #test set的檔案名稱

    # 確保資料夾存在
    os.makedirs(output_folder, exist_ok=True)

    # 打開輸出文件以寫入
    with open(output_file, 'w', encoding='utf-8') as outfile:
        # 遍歷每個文件
        for file_name in file_names:
            # 打開每個文件以讀取
            with open(file_name, 'r', encoding='utf-8') as infile:
                # 讀取內容並寫入輸出文件
                outfile.write(infile.read())
                #outfile.write('\n')  # 添加換行符以分隔文件內容

    # 複製 S5.txt 並重命名為 test.txt
    shutil.copy(source_file, destination_file)

    print(f'已成功合併文件到 {output_file}')
    print(f'已成功複製 {source_file} 到 {destination_file}')
    




# 定義要合併的文件名
# file_names = ['S1.txt', 'S2.txt', 'S3.txt', 'S4.txt'] #要合併的txt，作為training set
# output_folder = 'Fold1' #要存的資料夾
# output_file = os.path.join(output_folder, 'training.txt') #training set的檔案名稱
# source_file = 'S5.txt' #作為test set的檔案
# destination_file = os.path.join(output_folder, 'test.txt') #test set的檔案名稱

# # 確保資料夾存在
# os.makedirs(output_folder, exist_ok=True)

# # 打開輸出文件以寫入
# with open(output_file, 'w', encoding='utf-8') as outfile:
#     # 遍歷每個文件
#     for file_name in file_names:
#         # 打開每個文件以讀取
#         with open(file_name, 'r', encoding='utf-8') as infile:
#             # 讀取內容並寫入輸出文件
#             outfile.write(infile.read())
#             #outfile.write('\n')  # 添加換行符以分隔文件內容

# # 複製 S5.txt 並重命名為 test.txt
# shutil.copy(source_file, destination_file)

# print(f'已成功合併文件到 {output_file}')
# print(f'已成功複製 {source_file} 到 {destination_file}')
```
