```python
import pandas as pd
import openpyxl
# 變數設定，目的在於其他的project只要修改這部分就好了
Project = 'time'
Project2 = 'Time'
project_bugid_num = 27
Excel_name = 'time_整理結果'
Excel_save_path = 'time_result'

# 這是要放要拿來使用的絕對路徑
# unsort_data_path = 'Z:/廷偉/偉/研究所/其他SBFL資料-用於後續實驗用/CalculateSpectrum/CalculateSpectrum/CalculateSpectrum/Chart/Sus/old'
unsort_data_path = '懷疑值source/' + Project2 + '/old'

SBFL_all = ["Ochiai"]
```

```python
# 取得project的每一個bug-id有多少行的方法。
SBFL_fault_statement_number = []
project_statement_file = 'project_line/' + Project +'_line.txt'

# 使用 with 語句打開文件，確保文件在使用後自動關閉
with open(project_statement_file, 'r', encoding='utf-8') as file:
    project_lines = file.readlines()  # 讀取所有行

# 去掉每行末尾的換行符
project_lines = [int(line.strip()) for line in project_lines] # 順便將str轉為數字
print(project_lines)

max_line = max(project_lines) # 找出最大行數，用於後續輸出使用
print(max_line)
```

```python
# get_fault_statement
#動態產生二微陣列的方法
Project_bugid_faultStatement = []

for i in range(0,project_bugid_num):
    fault_statement_file_path = '錯誤位置/'+ Project + '/' + Project2 +'_v' + str(i+1) + '_FaultStatement.txt'
    print(fault_statement_file_path)

    # 使用 with 語句打開文件，確保文件在使用後自動關閉
    with open(fault_statement_file_path, 'r', encoding='utf-8') as file:
        lines2 = file.readlines()  # 讀取所有行

    # 去掉每行末尾的換行符
    lines2 = [line.strip() for line in lines2]
    print(lines2)

    lines2 = lines2[0].split()
    print(len(lines2))

    Project_bugid_faultStatement.append([])
    for j in range(0,len(lines2)):
        Project_bugid_faultStatement[i].append(int(lines2[j])) # 順便將str轉為數字

print(Project_bugid_faultStatement)

```

```python
# #要存成的excel index有甚麼
# data = {
#     'SBFL_Tool': [],
#     'Top': [],
#     'Spectra_file_index': [],
#     'Suspicious_Value': []
# }
```

```python
#動態產生二微陣列的方法
bug_id_rank = []

for SBFL_index, SBFL_a in enumerate(SBFL_all):
    
    
    for i in range(0,project_bugid_num):   

        # 要存成的excel index有甚麼
        # 因為每次一個for迴圈就會要新的df，所以每次都要重新定義，保證能清空前一個錯誤版本的資料
        data = {
        'SBFL_Tool': [],
        'bug_id': [],
        'Top': [],
        'Spectra_file_index': [],
        'Suspicious_Value': [],
        '錯誤': []
        }

        # 後續記錄用
        # 紀錄不同錯誤版本用於SBFL後，錯誤在哪個排名
        bug_id_rank.append([])

        file_path = unsort_data_path + '/' + SBFL_a + '_v' + str(i+1) + '_sus.txt' #使用絕對路徑
        print("懷疑值檔案", file_path)

        # 使用 with 語句打開文件，確保文件在使用後自動關閉
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()  # 讀取所有行

        # 去掉每行末尾的換行符
        lines = [line.strip() for line in lines]

        # 轉換為float，以及處理例外的問題。
        numbers = []
        for line in lines:
            try:
                number = float(line.strip())
            except ValueError:  # 捕獲轉換錯誤
                print("抓到非數字!")
                number = 0  # 將無法轉換的值設為0
            numbers.append(number)


        # 將列表中的元素和其索引配對，用於表示的幾個statement
        indexed_list = list(enumerate(numbers))

        # 取出最大的數字
        max_prob = max(indexed_list)
        count = numbers.count(max_prob)

        # 將列表中的元素和其索引配對,索引從1開始
        indexed_list = [(i+1, numbers) for i, numbers in enumerate(numbers)]

        # 找出前 XX 名（若列表長度小於 XX，則返回所有元素）
        # top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:max_line]

        # 找出前 200 名（若列表長度小於 200，則返回所有元素），因為全部都存起來，excel會爆炸，所以改成200名就好了。
        #top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:200]

        # 因為不需要前200名，所以取全部。
        top_test = sorted(indexed_list, key=lambda x: x[1], reverse=True)

        Now_top = 0 # 用於紀錄top用的
        previous_rank_sus = '' #表示與前一個rank的懷疑質做比較做比較
        count_top = 0

        for index, value in top_test:

            #用於處理計算Top使用----
            count_top += 1
            if value != previous_rank_sus:
                #Now_top = Now_top
                #Now_top += 1
                Now_top = count_top
            else:
                Now_top = Now_top
            #----------------------
            
            # 標示錯誤位置用
            flag = 0
            for faultStatement_index in range(0,len(Project_bugid_faultStatement[i])): # 檢查這一行有沒有存在正確的錯誤位置(Project_bugid_faultStatement)裡面。
                if index == Project_bugid_faultStatement[i][faultStatement_index]:
                    bug_id_rank[i].append("錯誤index: " + str(index) + ", Top位置: " + str(Now_top)) # 順便將str轉為數字
                    flag = 1
                    break

            previous_rank_sus = value
            data['SBFL_Tool'].append(SBFL_a)
            data['bug_id'].append(i+1)
            data['Top'].append(Now_top)
            data['Spectra_file_index'].append(index)
            data['Suspicious_Value'].append(value)
            if  flag == 1:
                data['錯誤'].append("錯誤找到")
            elif flag == 0:
                data['錯誤'].append(" ")

        # 創建 DataFrame
        df = pd.DataFrame(data)
        print(df)

        save_excel_file_name = Excel_name + SBFL_a + '_' + Project + '_v' + str(i+1)
        print(save_excel_file_name)
        wb = openpyxl.Workbook() #建立空的excel
        #wb = openpyxl.load_workbook('oxxo.xlsx') #開啟excel
        wb.save( Excel_save_path + '/' + save_excel_file_name + '.xlsx')

        # # 將資料寫入 Excel 檔案
        file_name = Excel_save_path + '/' + save_excel_file_name + '.xlsx'
        df.to_excel(Excel_save_path + '/' + save_excel_file_name + '.xlsx', index=False)

        print(f"資料已寫入 {save_excel_file_name} 文件中")


print(bug_id_rank)
```
